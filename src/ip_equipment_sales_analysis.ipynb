```python
import pandas as pd
import numpy as np 
```

```python
df_ip_equipment_details = pd.read_excel('../data/processed/Ip_Equipment_Details.xlsx')
df_ip_equipment_sales = pd.read_excel('../data/processed/Ip_Equipment_Sales.xlsx')
```

```python
df_ip_equipment_details.head()
```

```python
df_ip_equipment_details= df_ip_equipment_details.drop(['Unnamed: 0'],axis=1)
```

```python
df_ip_equipment_details.head()
```

```python

df_ip_equipment_sales.head()
```

```python
df_ip_equipment_sales=df_ip_equipment_sales.drop(['Unnamed: 0'],axis=1)
```

```python
df_ip_equipment_sales.head()
```

```python
df_ip_equipment_final_report_data = pd.merge(df_ip_equipment_sales,df_ip_equipment_details,left_index=True,right_index=True)
```

```python
df_ip_equipment_final_report_data.head()
```

```python
df_ip_equipment_final_report = df_ip_equipment_final_report_data.rename(columns={'(No column name)':'Department'})
```

```python
df_ip_equipment_final_report.head()
```

```python
df_ip_equipment_final=df_ip_equipment_final_report.T.drop_duplicates().T
```

```python
df_ip_equipment_final = df_ip_equipment_final.rename(columns={'type':'Test Type'})
```

```python
df_ip_equipment_final.head()
```

```python
df_ip_equipment_final.to_excel('../data/final/Ip_Equipment_Final.xlsx')
```
